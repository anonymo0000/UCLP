```python
from nltk.corpus import wordnet as wn

#       ，  Wu-Palmer          
def sort_by_similarity(word):
    #              
    synsets = wn.synsets(word)
    #         ，     
    if not synsets:
        return []
    #   ，             
    else:
        reference = synsets[0]
    #        ，               
    result = []
    #         
    for synset in synsets:
        #             Wu-Palmer   
        similarity = reference.wup_similarity(synset)
        #        None，                  
        if similarity is not None:
            result.append((synset, similarity))
    #              
    result.sort(key=lambda x: x[1], reverse=True)
    #         10          
    return [x[0] for x in result[:30]]

#     
word = 'argument'
synsets = sort_by_similarity(word)
#     
print(f'    10      ：')
for synset in synsets:
    print(synset)
```
